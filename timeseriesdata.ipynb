```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={"Data":pd.to_datetime(["2022-01-01","2022-02-02","2022-03-01","2022-03-02","2022-03-04","2022-04-01","2022-04-15",'2022-04-30',"2022-05-05","2022-06-06","2022-07-07","2022-08-08","2022-09-09","2022-10-10","2022-11-11"]),
"value":[10,15,22,28,35,42,48,35,60,68,75,82,90,90,105]}
df=pd.DataFrame(data)
df.set_index("Data",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker='o')
plt.xlabel("Data")
plt.ylabel("value")
plt.title("Time series with trend")
plt.grid(True)
plt.show()
```
