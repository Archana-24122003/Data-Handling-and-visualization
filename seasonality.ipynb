```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "Data":pd.to_datetime(["2020-01-01","2020-06-06","2020-07-07","2022-02-02","2022-11-11","2022-12-12","2023-01-01","2023-02-02","2023-03-03","2023-04-04",
                           "2024-05-05","2024-06-06","2024-07-07"]),
    "value":[0,10,20,30,40,50,60,75,40,40,41,50,60,]
}
df=pd.DataFrame(data)
df.set_index("Data",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"])
plt.xlabel("Data")
plt.ylabel("value")
plt.title("Time series with trend ")
plt.grid(True)
plt.show()
```
